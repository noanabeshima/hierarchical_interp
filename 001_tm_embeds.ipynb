```python
import tinymodel


lm = tinymodel.TinyModel()
```

```python
embeds = lm.embed.weight.data
```

```python
import tokre

tokre.setup(tokenizer=tinymodel.tokenizer, workspace='workspace')
```

```python
import torch
name_toks = tokre.load_literal_set('names')['literal_toks']
names = [name[0] for name in name_toks if len(name) == 1 and ''.join(name).strip()[0].isupper() and name[0][0]==' ']

name_ids = [tokre.enc(name)[0] for name in names]

# Create a mask for name tokens
name_mask = torch.zeros(embed.shape[0], dtype=torch.bool)
name_mask[name_ids] = True

name_embeds = embeds[name_ids]
# Calculate the mean embedding for names
name_mean_embed = embeds[name_mask].mean(dim=0)

# Calculate the difference between name mean and overall mean
name_dir = name_mean_embed - mean_embed

# Normalize the name direction vector
name_dir = name_dir / torch.norm(name_dir)
```

```python
from noa_tools import hist

name_weights = (embeds @ name_dir)
hist(name_weights, info=tinymodel.raw_toks)
```
